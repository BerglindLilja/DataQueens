Provide an example of a regex to match 4 digits numbers (by this, I mean precisely 4 digits, you should not match any part of numbers with e.g. 5 digits). In your notebook, use findall to show that your regex works on this test-text. Hint: a great place to test out regular expressions is: https://regex101.com.


```python
str = '12345 1234'
match = re.search(r'\d\d\d\d', str)
# If-statement after search() tests if it succeeded
if match:
  print( 'found', match.group()) ## 'found word:cat'
else:
  print ('did not find')
```

Provide an example of a regex to match words starting with "super". Show that it works on the test-text.


```python
url = 'https://raw.githubusercontent.com/SocialComplexityLab/socialgraphs2021/main/files/regex_exercise.txt'
response = urllib.request.urlopen(url)
data = response.read()      # a `bytes` object
text = data.decode('utf-8')
#print(text)
```

```python
match2 = re.findall(r'super\w*', text)
# If-statement after search() tests if it succeeded
if match2:
  print( 'found', match2) ## 'found word:cat'
else:
  print ('did not find')
```

```python
zeldaurl = 'https://raw.githubusercontent.com/SocialComplexityLab/socialgraphs2021/main/files/Zelda_regex_exercise.txt'
zeldaresponse = urllib.request.urlopen(zeldaurl)
zeldadata = zeldaresponse.read()      # a `bytes` object
zeldatext = zeldadata.decode('utf-8')
#print(zeldatext)
```

```python
match3 = re.findall(r'\[\w*\s*\w*\]', zeldatext)
# If-statement after search() tests if it succeeded
if match3:
  print( 'found', match3) ## 'found word:cat'
else:
  print ('did not find')
```

## Part A, download zelda wiki pages

```python
import urllib.request
baseurl = 'https://zelda.fandom.com/api.php?'
action = "action=query"
title = "titles=Characters_in_Breath_of_the_Wild"
content = "prop=revisions&rvprop=content&rvslots=*"
dataformat ="format=json"

query1 = "{}{}&{}&{}&{}".format(baseurl, action, content, title, dataformat)
#query1 = "{}{}?{}&{}".format(baseurl, title, content,dataformat)
print(query1)
```

```python
import json

wikiresponse1 = urllib.request.urlopen(query1)
wikidata1 = wikiresponse1.read()
wikitext1 = wikidata1.decode('utf-8')
#print(wikitext1)
charecters = json.loads(wikitext1)
charectersContent = charecters['query']['pages']['77604']['revisions'][0]['slots']['main']['*']
# print(charectersContent)
#print('Great success')
```

```python
print(type(charectersContent))
```

```python
title2 = "titles=Enemies_in_Breath_of_the_Wild"
query2 = "{}{}&{}&{}&{}".format(baseurl, action, content, title2, dataformat)
print(query2)
```

```python
wikiresponse2 = urllib.request.urlopen(query2)
wikidata2 = wikiresponse2.read()
wikitext2 = wikidata2.decode('utf-8')
enemies = json.loads(wikitext2)
print(enemies)
enemiesContent = enemies['query']['pages']['77608']['revisions'][0]['slots']['main']['*']
print(enemiesContent)
```

```python
title3 = "titles=Bosses_in_Breath_of_the_Wild"
query3 = "{}{}&{}&{}&{}".format(baseurl, action, content, title3, dataformat)
print(query3)
```

```python
wikiresponse3 = urllib.request.urlopen(query3)
wikidata3 = wikiresponse3.read()
wikitext3 = wikidata3.decode('utf-8')
bosses = json.loads(wikitext3)
# print(bosses)
bossesContent = bosses['query']['pages']['77607']['revisions'][0]['slots']['main']['*']
print(bossesContent)
```

```python
import pandas as pd
df = pd.DataFrame(columns={"name","type"})
```

```python

match4 = re.findall(r'\{\{Term\|BotW\|.*\}\}', charectersContent)
# If-statement after search() tests if it succeeded
if match4:
  # print( 'found', match4) ## 'found word:cat'
  for item in match4:
    print (item)
else:
  print ('did not find')
```

```python
match4 = re.findall(r'(\{\{Term\|BotW\|)(.*)(\|link\}\})', charectersContent)# If-statement after search() tests if it succeeded
if match4:
  # print( 'found', match4) ## 'found word:cat'
  for item in match4:
    print (item[1])
else:
  print ('did not find')
```

```python
regDf =pd.DataFrame(columns=['Name', 'type'])
for item in match4:
    # regDf.append([item[1],'Ally'])
    regDf.loc[len(regDf.index)] = [item[1],'Ally'] 
print(regDf.head())
print(regDf.tail())


  
# display(df)
```

```python
match5 = re.findall(r'(\{\{Term\|BotW\|)(\w*.\w*)(\|link\}\})', enemiesContent)# If-statement after search() tests if it succeeded
if match5:
#   print( 'found', match4) ## 'found word:cat'
  for item in match5:
    print (item)
else:
  print ('did not find')
```

```python
enemyDf =pd.DataFrame(columns=['Name', 'type'])
for item in match5:
    # regDf.append([item[1],'Ally'])
    enemyDf.loc[len(enemyDf.index)] = [item[1],'Enemy'] 
print(enemyDf.head())
print(enemyDf.tail())


  
# display(df)
```

```python
match6 = re.findall(r'(\{\{Term\|BotW\|)(\w*.\w*)(\|link\}\})', bossesContent)# If-statement after search() tests if it succeeded
if match6:
#   print( 'found', match4) ## 'found word:cat'
  for item in match6:
    print (item)
else:
  print ('did not find')
```

```python
bossDf =pd.DataFrame(columns=['Name', 'type'])
for item in match6:
    # regDf.append([item[1],'Ally'])
    bossDf.loc[len(bossDf.index)] = [item[1],'Boss'] 
print(bossDf.head())
print(bossDf.tail())
```

```python
# allDf = pd.DataFrame(regDf,enemyDf,bossDf)
allDf = pd.DataFrame(columns=['Name', 'type'])
allDf = allDf.append(regDf,ignore_index = True)
allDf = allDf.append(enemyDf,ignore_index = True)
allDf = allDf.append(bossDf,ignore_index = True)
print(allDf.head())
print(allDf.tail())
```
