### Binning and plotting degree distributions

```python
N = 5000
k = 10
p = k/(N-1)
print(p)
```

```python
g = nx.erdos_renyi_graph(N,p)

nx.draw_shell(g)
```

```python
def plot_degree_dist(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()
    
# naming the x axis
plt.xlabel('Degree distribution')

# naming the y axis
plt.ylabel('# of numbers in each bin')
 
# graph title
plt.title('Erd√∂s-Renyi network graph')

plot_degree_dist(g)
```
