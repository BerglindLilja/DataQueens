```python
import re
import urllib.request
import pandas as pd
import json
import urllib.parse
```

```python
charactersDf = pd.read_csv('characters.csv')
charactersDf.head()
```

```python
baseurl = 'https://lotr.fandom.com/api.php?'
action = "action=query"
title = "titles="
content = "prop=revisions&rvprop=content&rvslots=*"
dataformat ="format=json"
character = 'Théoden'
character = urllib.parse.quote(character)

query1 = "{}{}&{}&{}&{}".format(baseurl, action, content, title + character, dataformat)

wikiresponse1 = urllib.request.urlopen(query1)
wikidata1 = wikiresponse1.read()
wikitext1 = wikidata1.decode('utf-8')
response = json.loads(wikitext1)
pageid =list(response['query']['pages'].keys())[0]
responseContent = response['query']['pages'][pageid]['revisions'][0]['slots']['main']['*']
print(responseContent)
```

```python
baseurl = 'https://lotr.fandom.com/api.php?'
action = "action=query"
title = "titles="
content = "prop=revisions&rvprop=content&rvslots=*"
dataformat ="format=json"

for name in charactersDf.Name:
    try:
        tmp = name.replace(' ','_')
        tmp = urllib.parse.quote(tmp)
        query1 = "{}{}&{}&{}&{}".format(baseurl, action, content, title + tmp , dataformat)
        wikiresponse1 = urllib.request.urlopen(query1)
        wikidata1 = wikiresponse1.read()
        wikitext1 = wikidata1.decode('utf-8')
        response = json.loads(wikitext1)
        pageid =list(response['query']['pages'].keys())[0]
        responseContent = response['query']['pages'][pageid]['revisions'][0]['slots']['main']['*']
        text_file = open("txtfiles/" + name + ".txt", "wt", encoding="utf-8")
        n = text_file.write(responseContent)
        text_file.close()
        print(tmp)
    except:
        print('oops', name)
```
