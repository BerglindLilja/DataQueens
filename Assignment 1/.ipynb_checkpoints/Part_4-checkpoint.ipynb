```python
import networkx as nx
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def BAGraph(size):   
    tmp = nx.Graph()
    tmp.add_nodes_from([1,2])
    tmp.add_edge(1,2)
    #Create a list of all the edges in the graph and flatten the list
    edges = list(tmp.edges)
    flat_list = [item for sublist in edges for item in sublist]

    for i in range(3,size+1):
        edges = list(tmp.edges)
        flat_list = [item for sublist in edges for item in sublist]
        x = random.choice(flat_list)
        tmp.add_node(i)
        tmp.add_edge(i,x)
    return tmp
   

        
```

```python
G = BAGraph(100)
nx.draw(G, node_size = 50)
```

```python
G5000 = BAGraph(5000)
```

```python
print("max",max(G5000.degree, key=lambda x: x[1])[1])
print("min",min(G5000.degree, key=lambda x: x[1])[1])
```

```python
degrees = [G5000.degree(n) for n in G5000.nodes()]
#print(degrees)
#print(np.histogram(degrees))
hist,bin_edges = np.histogram(degrees,100)
print(bin_edges)
print(hist)

plt.scatter(bin_edges[:-1],hist)
plt.xlabel("k")
plt.ylabel("Count")
```

```python
fig, ax = plt.subplots()
  
ax.set_yscale("log")
ax.scatter(bin_edges[:-1],hist)
#ax.plot([1, 3, 10], [1, 9, 100], "o-",color ="green")
  
ax.set_ylim(1E-1, 1e5)
#ax.set_aspect(1)
plt.xlabel("k")
plt.ylabel("Count")
plt.show()
```
